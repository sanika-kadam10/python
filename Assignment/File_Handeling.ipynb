7. Write a python program to find the longest words.


```python
with open("demo.text","r") as f:
  data=f.read()
words=data.split()
longest=""
for i in words:
  if len(i)>len(longest):
    longest = i
print("Longest word is:",longest)
```

9. Write a Python program to count the frequency of words in a file.


```python
with open("demo.text","r") as f:
  data=f.read()
words=data.split()
count={}
for i in words:
  if i in count:
    count[i] +=1
  else:
    count[i] = 1
for i in count:
  print(count)
  print(i,":",count[i])
```
